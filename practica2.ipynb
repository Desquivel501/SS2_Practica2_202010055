```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('datos.csv')
df
```

```python
# Limpiar datos
df['Rating'] = df['Rating'].fillna(0)
df['What you will learn'] = df['What you will learn'].fillna("")
df['Duration'] = df['Duration'].fillna("")
df['Schedule'] = df['Schedule'].fillna("")
df['Review'] = df['Review'].fillna("0 reviews")
df
```

###  Promedio de calificaciones

```python
rating_avg = pd.Series(df['Rating']).groupby(df['Course Title']).mean().reset_index()
rating_avg.to_csv('output/rating_avg.csv', index=False)
rating_avg
```

Al tener el valor promedio de entre todas las reviews registradas se podra realizar un analisis sobre estas.

### Cursos con mayor y menor rating

```python
top_rating_avg = rating_avg.sort_values(by='Rating', ascending=False).head(10)
top_rating_avg.to_csv('output/top_rating_avg.csv', index=False)
top_rating_avg
```

EN este bloque se listan los cursos con mejor rating, estos no parecen tener un tema comun, sienfo algunos de gestion de proyectos, otros de enseñanza y otros de música.

```python
bot_rating_avg = rating_avg.sort_values(by='Rating', ascending=True).head(10)
bot_rating_avg.to_csv('output/bot_rating_avg.csv', index=False)
bot_rating_avg
```

En este bloque podemos ver algunos de los cursos con menor rating. De estos vemos que la mayoria son relacionados a finanzas y negocios, por lo que se podria concluir que estos son los cursos menos populares en el sistema.

### Porcentaje de cursos con horario flexible

```python
horario_flex = df[df['Schedule'].str.contains('Flexible', case=False)]
percentage = (horario_flex.shape[0] / df.shape[0]) * 100
print(str(round(percentage, 2)) + "%")
```

De este porcentaje se puede concluir que casi todos los cursos ofrecen un horario flexible. Est podria ser debido a que estos cursos se realizan por videos on-demand, por lo que el usuario es libre de realizar el curso a su propio ritmo.

### Número de cursos por nivel de dificultad

```python
plt.figure(figsize=(8, 6))
df.groupby('Level').size().plot(kind='bar')
plt.title('Número de cursos por nivel')
plt.xlabel('Nivel')
plt.ylabel('Cantidad de cursos')
plt.xticks(rotation=90)
plt.show()
```

En esta grafica se puede apreciar que la mayoria de los cursos que se ofrecen en el sistema son para principiantes, seguidos por los de nivel intermedio y por ultimo un bajo numero de cursos avanzados. Esto se puede deber a que muchas personas empiezan cursos avanzados pero no continuan con los demas niveles, por lo que no resulta rentable tener tantos cursos de niveles intermedios y avanzados.

### Número de cursos por categoria

```python
plt.figure(figsize=(8, 6))
df.groupby('Keyword').size().plot(kind='barh', color='green')
plt.title('Número de cursos por categoria')
plt.ylabel('Categoria')
plt.xlabel('Cantidad de cursos')
plt.xticks(rotation=90)
plt.show()
```

En el grafico se observa que una gran parte de los cursos que se ofrecen son de ciencias sociales (1800), teniendo una gran diferencie en comparación de los demas. Los demas cursos tienen una cantidad similar de cursos (1000) a excepcion de "artes y humanidades" y "matematica y logica", que tienen 500 cada uno.

### Relación entre la duración del curso y el número de reviews

```python
# num_reviews = pd.Series(df['Rating']).groupby(df['Course Title']).count().reset_index()

# result = pd.merge(num_reviews, df, on='Course Title')
# result
import re

def extract_hours(duration):
    
    pattern = r'\b(\d+)\b'
    matches = re.findall(pattern, duration)
    
    if len(matches) == 0:
        return 0
    elif len(matches) == 1:
        return int(matches[0])
    else:
        return int(matches[0]) * 4 * int(matches[1])
    
def extract_reviews(review):
    
    pattern = r'\b(\d+)\b'
    matches = re.findall(pattern, review)
    
    if len(matches) == 0:
        return 0
    else:
        return int(matches[0])
    

df['Hours'] = df['Duration'].apply(extract_hours)
df['Reviews'] = df['Review'].apply(extract_reviews)

plt.figure(figsize=(8, 6))
plt.scatter(df['Hours'], df['Reviews'], color='red')
plt.title('Número de duración vs reviews')
plt.xlabel('Duración')
plt.ylabel('Número de reviews')
plt.show()
```

Se puede observar que no hay una relacion muy clara entre el numero de reviews y la duración. Tenemos que la mayoria de cursos del sistema son de menos de 100 horas, entre estos el numero de reviews es bastante disperso, con una leve inclinacion a tener pocas reviews. Sin embargo, mientras los cursos son mas largos notamos que estos tienen a tener menos reviews. Por lo que se podria concluir que los usuarios son mas propensos a iniciar, terminar y calificar cursos de menor duración que cursos de larga duración.

### Histograma de la distribución de las duraciones de los cursos

```python
plt.figure(figsize=(8, 6))
df.plot(kind='hist', y='Hours', bins=30, color='purple')
plt.title('Distribución de duración de cursos')
plt.xlabel('Duración')
plt.ylabel('Cantidad')
plt.show()
```

Como se menciono anteriromente, podemos ver que casi todos los cursos que se ofrecen tienen una duración menor a 50 horas, siendo muy pocos los que duran mas de 100.

### Distribución de las calificaciones de los cursos por nivel de dificultad

```python
advanced = df[df['Level'] == 'Advanced level']
intermediate = df[df['Level'] == 'Intermediate level']
beginner = df[df['Level'] == 'Beginner level']

plt.figure(figsize=(8, 6))
plt.boxplot([advanced['Rating'], intermediate['Rating'], beginner['Rating']], labels=['Advanced', 'Intermediate', 'Beginner'])
plt.title('Boxplot de rating por nivel de dificultad')
plt.ylabel('Rating')
plt.show()
```

En este grafico se ve que a paesar de algunos valores atipicos, los ratings de  los niveles avanzados e intermedios se comportan de una manera similar, encontrandose mayormente entre los mediados de los 4s. El nivel para principiantes es un poco distinto, este tiene mas valores atipicos que los otros dos niveles combinados, pero la mayoria se encuentra por encima de 4.5 con una varianza bastante pequeña. Esto puede significar que ñor lo general los cursos de prinicpiante son mejor calificados que los cursos intermedios y avanzados.
